<h1>This program scrapes and summarizes news articles</h1>

```
import nltk 
from newspaper import Article
import datetime as dt
```

```
#Get the article

url = 'https://www.washingtonpost.com/technology/2019/07/17/you-downloaded-faceapp-heres-what-youve-just-done-your-privacy/?noredirect=on&utm_term=.1938589d078f'

article = Article(url)
```

```
#NLP
article.download()
article.parse()
nltk.download('punkt')
article.nlp()
```

```
#Get the author(s)
article.authors
```

```
#Get the published date
article.publish_date.strftime('%d/%m/%y')
```

```
#Top image
article.top_image
```

```
#Get the article text
article.text
```

```
#Get a summary of the article
article.summary
```
