```
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```
#Download stopwords package
nltk.download('stopwords')
```

```
def process_text(text):

  #Remove punctuation, stopwords
  no_punc = [char for char in text if char not in string.punctuation]
  no_punc = ''.join(no_punc)

  #Return list of clean text words
  clean_words = [word for word in no_punc.split() if word.lower() not in stopwords.words('english')]

  return clean_words
```

```
#Tokenization

df['text'].head().apply(process_text)
```

```
message4 = 'hello world hello hello world play'
message5 = 'test test test test one hello'

#bow = bag of words
#Convert text to matrix of token counts
bow4 = CountVectorizer(analyzer=process_text).fit_transform([[message4],[message5]])

#Unique words in sentences (2nd column)
print(bow4)

print(bow4.shape)
```

```
messages_bow = CountVectorizer(analyzer=process_text).fit_transform(df['text'])
```

```
#Split the data 80% Training, 20% Testing
#Random state = 0 allows different results
X_train,X_test,y_train,y_test = train_test_split(messages_bow, df['spam'],test_size=0.2,random_state=0)
```

```
messages_bow.shape
```

```
#Create and train Naive Bayes
classifier = MultinomialNB()

classifier.fit(X_train,y_train)
```

<h1>Predictions on training data</h1>


```
#Predictions on training data
print(classifier.predict(X_train))

#Actual target values
print(y_train.values)
```

```
#Evaluate model on training data set

prediction = classifier.predict(X_train)
print(classification_report(y_train,prediction))
```

```
#Confusion Matrix
print(confusion_matrix(y_train,prediction))
```

```
#Accuracy on training data 99.7%
print(accuracy_score(y_train,prediction))
```
