<h1>Chatbot Program</h1>


```
#Import modules
from nltk.chat.util import Chat, reflections
```

```
#Create a single asnwer-response pair
pairs = [
         ["My name is Bob", ['Hi Bob']]
        ]

#Regex Pattern
Pairs = [
         ['My name is (.*)', ['Hi %1']],
         ['(hi|hello|hey|holla|hola)', ['hey there','hi there','haaaayyyyy']],     # | means "or"; same output response(s) to any of the given inputs
         ['(.*) in (.*) is fun', ['%1 in %2 is indeed fun']], # %1 corresponds to first value, %2 corresponds to second value, so on
         ['(.*)(location|city) ?', 'Tokyo, Japan'],
         ['(.*) created you?', ['Someone did using NLTK']],
         ['How is the weather in (.*)', ['The weather in %1 is amazing like always']],
         ['(.*)help(.*)', ['I can help you']], # If the input contains anything within (.*) (.*)
         ["(.*)your name?",['My name is Jarvis']]
        ]
```

```
chat = Chat(Pairs, reflections)

#Create a substitue response string
#chat._substitute('You are amazing!')

chat.converse()
```

```
#Input values with output values in reflections
print(reflections)
```

```
my_dummy_reflections = {
                        'go':'gone',
                        'hello':'hey there'
                       }
```

```
chat = Chat(pairs, my_dummy_reflections)
chat._substitute('You are amazing!')
```
