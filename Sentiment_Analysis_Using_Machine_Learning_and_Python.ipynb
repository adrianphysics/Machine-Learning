<h1>Sentiment of text from website</h1>

```
from textblob import TextBlob
import nltk
from newspaper import Article
```

```
#Get the article
url = 'https://everythingcomputerscience.com'
article = Article(url)
```

```
#NLP
article.download()
article.parse()

nltk.download('punkt')
article.nlp()
```

```
#Summary of the article
text = article.summary

text
```

```
#Create a TextBlob object
obj = TextBlob(text)

#Returns value between -1 and 1; -1 negative, 0 neutral, 1 positive
sentiment = obj.sentiment.polarity

if sentiment == 0:
  print('Neutral')
elif sentiment == 1:
  print('Positive')
elif sentiment == -1:
  print('Negative')
```
