<h1>Bitcoin price prediction</h1>

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```
df = pd.read_csv('BitcoinPrice.csv')
```

```
df.head()
```

```
#Remove date column

df.drop(['Date'],axis=1,inplace=True)
```

```
df.head()
```

```
#how many days to predict
prediction_days = 30

#Create another column shifted n units up
df['Prediction'] = df[['Price']].shift(-prediction_days)
```

```
df.head()
```

```
#Split data into independent and dependent datasets

#Convert dataframe to a numpy array and remove the prediction column
X = np.array(df.drop(['Prediction'],axis=1))
```

```
#Remove last 'n' rows (n = prediction days)
X = X[:len(df)-prediction_days]
```

```
print(X)
```

```
#Create the dependent data set

#Convert the dataframe to a numpy array
y = np.array(df['Prediction'])

#Get all of the values except for last 'n' rows
y = y[:-prediction_days]
```

```
#80% train, 20% test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```
#Set the prediction days array equal to the last 30 rows from the original data set
prediction_days_array = np.array(df.drop(['Prediction'],axis=1))[-prediction_days:]

prediction_days_array
```

<h1>SVM</h1>

```
#Create and train the SVM (regression) using radial basis function (rbf)

svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(X_train,y_train)
```

```
#Evaluate the model's accuracy
svr_rbf_confidence = svr_rbf.score(X_test,y_test)
print(svr_rbf_confidence)
```

```
#Predicted values of bitcoin
#Model using X_test to predict y_test

svm_prediction = svr_rbf.predict(X_test)

print(svm_prediction)

#Actual
print(y_test)
```

```
#Model predictions for the next n=30 days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)

#Actual values of bitcoin last 30 days
print(df.tail(prediction_days))
```
