<h1>Program determines how similar resume is to job description</h1>


```
import docx2txt
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```
#Load data
from google.colab import files
uploaded = files.upload()
```

<h1>Resume</h1>

```
resume = docx2txt.process('python_resume.docx')

resume
```

<h1>Job Description</h1>

```
job_description = docx2txt.process('job_description.docx')

job_description
```

```
text = [resume, job_description]
```

```
cv = CountVectorizer()

count_matrix = cv.fit_transform(text)
```

```
#64% similar
print(cosine_similarity(count_matrix))
```

```
#Match percentage
match_percentage = cosine_similarity(count_matrix)[0][1] * 100

#Get the similarity percentage
print(str(np.round(match_percentage,2))+'%')
```
