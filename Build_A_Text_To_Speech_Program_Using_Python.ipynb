<h1>Program takes text from online article and converts it to speech</h1>

```
from newspaper import Article
from gtts import gTTS
import nltk
import os
```

```
#Get the article
article = Article('https://hackernoon.com/future-of-python-language-bright-or-dull-uv41u3xwx')

#Download the article
article.download()

article.parse()
```

```
nltk.download('punkt')
```

```
#NLP

article.nlp()

#Get the articles text and store in variable
mytext = article.text

mytext
```

```
#Choose a language for the speech (english)
language = 'en'

#Convert text to speech
myobj = gTTS(text=mytext,lang=language,slow='False')
```

```
#Save converted audio
myobj.save('read_article.mp3')

#Play
os.system('start read_article.mp3')
```
